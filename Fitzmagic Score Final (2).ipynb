```python
import pandas as pd
```

```python
nfldata = pd.read_csv('qb_games.csv')
```

```python
nfldata.head()
```

```python
rating_overall = nfldata.loc[:,"Passing Rate"]
names = nfldata.loc[:,"NAME"]
```

```python
overall_stdev = rating_overall.std()
```

```python
overall_mean = rating_overall.mean()
```

```python
overall_mean
```

```python
print(overall_stdev)
```

```python
nfldata.describe()
```

```python
allqbs = set(names)
```

```python
stddev = nfldata.groupby('NAME')[['Passing Rate']].std()
```

```python
avg = nfldata.groupby('NAME')[['Passing Rate']].mean()
```

```python
combined = (stddev / overall_stdev) * (overall_mean / avg) * 100
```

```python
fitzmagic_score = combined.sort_values(by=['Passing Rate'],ascending = False)
```

```python
fitzmagic_score
```

```python
testdata = nfldata
```

```python
cleandata = testdata.groupby("NAME").filter(lambda x: len(x) >= 10)
```

```python
cleanstddev = cleandata.groupby('NAME')[['Passing Rate']].std()
```

```python
cleanavg = cleandata.groupby('NAME')[['Passing Rate']].mean()
```

```python
cleanavg
```

```python
finalscore = (cleanstddev / overall_stdev) * (overall_mean / cleanavg) * 100
```

```python
final_fitzmagic_score = finalscore.sort_values(by=['Passing Rate'],ascending = False)
```

```python
final_fitzmagic_score.to_csv("output2.csv",index=True)
```
